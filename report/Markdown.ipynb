3. Did strict schema mode prevent any bad output?
Answer:Partially. Strict schema mode helped ensure the model returned valid JSON with the expected structure, but it did not completely prevent incorrect values. During testing, the model still produced labels such as "billing_dispute" and "negative", which were not valid values in the Pydantic schema. The final validation layer (Pydantic) caught these issues and prevented them from being accepted. Therefore, strict schema mode improved output consistency, but Pydantic validation was still necessary to guarantee correctness.

4. What did you change in your prompt after the first run?
Answer:
After the first run, I made the prompt more explicit by:
Listing the exact allowed values for category, urgency, and sentiment.
Adding instructions such as:
"Do not create new labels."
"Do not use synonyms."
"Output only valid JSON."
Requiring all schema fields to be present in the response.
Specifying that if the model was uncertain, it should choose the closest valid label rather than inventing a new one.
These changes reduced schema validation errors and produced more consistent structured outputs.

I used a temperature of 0.2 because the support agent needed consistent and predictable outputs for classification, structured extraction, and tool-calling behavior. A low temperature reduced randomness and improved reliability.

The lock_account tool prevented the biggest possible mistake. It ensured that suspected fraud or account-takeover cases were escalated instead of being incorrectly processed as normal refund requests.

I would add a guardrail that automatically sets all fraud-risk tickets to critical urgency and blocks refund approval until a manual security review is completed.

I would monitor category classification accuracy, especially for fraud-related tickets, to ensure that high-risk cases are identified correctly. I would also track the percentage of tickets requiring manual correction or escalation.
