TEMP: 0
**Ticket Classification:**  
- **Category:** Billing – Duplicate/Double Charge  
- **Sub‑category:** Duplicate payment for order **QC‑1008** (two $89.99 charges)  

**Suggested Next Action:**  

1. **Acknowledge & Apologize**  
   - Respond promptly: “I’m sorry for the inconvenience. I see that you were charged twice for order QC‑1008.”  

2. **Verify the Charges**  
   - Pull the customer’s order details and payment logs for QC‑1008.  
   - Confirm that two separate $89.99 authorizations/settlements exist (date, time, transaction IDs).  

3. **Initiate Refund for the Duplicate**  
   - If the duplicate is confirmed, process a full refund for the extra $89.99.  
   - Provide the expected refund timeline (e.g., 3‑5 business days) and a reference number.  

4. **Update the Customer**  
   - Send a confirmation email/message with:  
     - Refund amount and transaction ID.  
     - Expected credit date.  
     - Contact info for any follow‑up questions.  

5. **Internal Follow‑up**  
   - Log the incident in the billing system as “Duplicate charge – QC‑1008.”  
   - Flag the payment gateway/processor for a quick review to prevent recurrence (possible double‑submission glitch).  

6. **Close the Ticket**  
   - Once the refund is processed and the customer confirms receipt, mark the ticket as **Resolved – Duplicate Charge Refunded**.  

**Template Response (for the support agent):**

```
Hi [Customer Name],

I’m sorry to hear about the duplicate charge on your account. I’ve reviewed your order QC‑1008 and can see that two $89.99 transactions were processed.

I’ve initiated a refund for the extra $89.99. You should see the credit appear on your bank statement within 3‑5 business days. I’ll send you a confirmation with the refund reference number shortly.

If you have any further questions or notice anything else, please let me know—I’m here to help.

Thank you for your patience,
[Your Name]
Customer Support
```

Following these steps will resolve the billing issue, restore the customer’s confidence, and ensure the problem is documented for future prevention.

**Ticket Classification**  
- **Category:** Billing – Duplicate Charge / Over‑charge  
- **Sub‑category:** Duplicate card transaction (same order ID, same amount)  

**Suggested Next Action**  

1. **Acknowledge & Apologize**  
   - Respond to the customer promptly (ideally within the next few hours) with a friendly apology:  
     *“I’m sorry you’re seeing two $89.99 charges for order QC‑1008. Let’s get this sorted out right away.”*

2. **Verify the Transaction**  
   - Pull up the order details for **QC‑1008** in the payment system.  
   - Check the payment gateway logs for:
     - Two separate authorization/settlement events for $89.99.
     - Any “re‑try” or “void” attempts that may have resulted in a duplicate capture.  
   - Confirm the customer’s bank statement (if needed) matches the timestamps.

3. **Determine the Root Cause**  
   - If the duplicate was caused by a system retry, network timeout, or manual entry error, note that for internal reporting.  
   - If the duplicate is a genuine error (e.g., two separate charges processed), flag it for the finance team.

4. **Process the Refund**  
   - Initiate a **full refund of one $89.99 charge** to the same payment method.  
   - Generate a refund reference number and include it in the reply to the customer.  
   - Ensure the refund is processed within the standard SLA (usually 1–2 business days) and inform the customer of the expected timeline for the funds to appear on their bank statement.

5. **Communicate the Resolution**  
   - Send a concise email/message:  
     ```
     Hi [Customer Name],

     Thank you for bringing this to our attention. We’ve confirmed that two $89.99 charges were applied to your card for order QC‑1008. One of those charges has been refunded (refund reference: REF123456). The refund should appear on your statement within 1–2 business days.

     We apologize for any inconvenience this may have caused. If you have any further questions or notice anything else, please let us know—we’re here to help!

     Best regards,
     [Your Name]
     Customer Support Team
     ```

6. **Close the Ticket**  
   - Mark the ticket as **Resolved – Refund Issued**.  
   - Add internal notes summarizing the investigation and refund details for future reference.

7. **Prevent Future Occurrences**  
   - Log the incident in the “Duplicate Charge” incident tracker.  
   - If the duplicate was due to a system retry or gateway glitch, forward the details to the engineering/operations team for a root‑cause analysis and a possible fix.  
   - Review the payment gateway’s retry/timeout settings to ensure they do not unintentionally create duplicate captures.
